# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.arange(100)
print (x)
y = x * 2
print (y)
z = x ** 2
print (z)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Your title")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )

ax.plot(x, y)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("Your title")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] ) # [ start_x, start_y, width, height ]

ax.plot(x, y)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("Your title")

ax_2 = fig.add_axes( [0.2, 0.5, 0.2, 0.3] )
ax_2.plot(x, z)

ax_3 = fig.add_axes( [1.2, 0., 1., 0.5] )
ax_3.scatter(x, z)

plt.show()
```

```python
z[20] = 550
z[21] = 350
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] ) # [ start_x, start_y, width, height ]

ax.plot(x, z)
ax.set_xlabel("X axis")
ax.set_ylabel("z axis")
ax.set_title("Your title")

ax_2 = fig.add_axes( [0.2, 0.5, 0.2, 0.3] )
ax_2.plot(x, z)
ax_2.set_xlim(18,25)
ax_2.set_ylim(300,700)
```

```python
fig, axes =  plt.subplots(3,3)
print(axes)
for row in axes:
    for ax in row: 
        ax.plot(x,y)
    row[1].plot(x,z, c='r')

plt.tight_layout()
```

```python
fig, axes =  plt.subplots(1,2)
print(axes)
axes[0].plot(x, y, c='b', ls='--', lw=3)
axes[1].plot(x, z, c='r', lw=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


ax = plt.figure().add_subplot(projection='3d')

X, Y = np.mgrid[0:6*np.pi:0.25, 0:4*np.pi:0.25]
Z = np.sqrt(np.abs(np.cos(X) + np.cos(Y)))

ax.plot_surface(X + 1e5, Y + 1e5, Z, cmap='autumn', cstride=2, rstride=2)

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_zlabel("Z label")
ax.set_zlim(0, 2)

plt.show()
```
