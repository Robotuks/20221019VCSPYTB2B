```python
def my_function(num_1, num_2):
    result = None
    if (num_1 + num_2) % 2 == 0:
        result = num_1 * num_2
    else:
        result = num_1 / num_2
    return result
```

```python
print (my_function(2,6) )
print (my_function(10,5) )
```

```python
my_function_variable = my_function

my_function_variable(20,3)
```

```python
def my_dict_function( *args ):
    my_dict = {}

    for i in range(1,11):
        my_dict[ i ] = 0
        print( my_dict )
        for number in args:
            if number % i == 0:
                my_dict[ i ] += 1

    return my_dict
```

```python
my_dict_function(1,2,65,30,31,22,7)
```

```python
import time

start_time = time.time()
print( 'start time:', start_time)

timeout = 5

while start_time + timeout > time.time():
    current_time = time.time()
    print ('time left:', start_time + 5 - current_time)
```

```python
def anomaly_checker(list_1):
    for i in range( len(list_1) - 1 ):
        # print (i)
        firstNum = list_1[i]
        secondNum = list_1[i+1]
        diff = secondNum - firstNum
        if diff == 1:
            print(f'There is no anomaly between number', firstNum, secondNum )
        else:
            print(f'There is anomaly between number', firstNum, secondNum )
```

```python
anomaly_checker([17,18,19,21])
```
